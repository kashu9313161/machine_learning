## API


```python
import pandas as pd 
import requests
```

```python
response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=...........")
```

```python
pd.DataFrame(response.json()['results'])
```

```python
# to get the list one by one 
for i in range(1,429):
    response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=...........&page="+str(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','vote_average','vote_count']]
    df = pd.concat([df,temp_df], ignore_index=True)
# code execute one by one and add the data to the dataframe and ignore the index and create new index for the new dataframe

# to convert into a CSV file
df.to_csv("top_rated_movies.csv", index=False)
```
