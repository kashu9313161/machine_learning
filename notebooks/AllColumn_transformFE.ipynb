```python
import numpy as np 
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('covid_toy.csv')
df.head()
# gender, city = nominal categorical data
# fever = simple imputer
# cough = ordinal encoder
```

```python
df.info()
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('has_covid', axis = 1)
                                                 ,df['has_covid'],test_size = 0.2)
x_train
```

## Long process doing operations one by one

```python
# adding simple imputer to fever col
si = SimpleImputer()
# all values of fever filled with the mean value
x_train_fever = si.fit_transform(x_train[['fever']])
# also the test data
x_test_fever = si.fit_transform(x_test[['fever']])
x_train_fever
```

```python
# ordinal encoder cough
oe = OrdinalEncoder(categories=[['Mild','Strong']])
x_train_cough = oe.fit_transform(x_train[['cough']])
x_test_cough = oe.fit_transform(x_test[['cough']])
x_train_cough
```

```python
ohe = OneHotEncoder(drop = 'first', sparse_output = False)
x_train_gender_city = ohe.fit_transform(x_train[['gender','city']])
x_test_gender_city = ohe.fit_transform(x_test[['gender','city']])
x_train_gender_city
```

```python
ohe.get_feature_names_out(['gender', 'city'])
```

```python
# x_train_age = df.drop(df[['gender','fever','cough','city','has_covid']], axis = 1)
x_train_age = x_train.drop(columns=['gender','fever','cough','city']).values
x_test_age = x_test.drop(columns=['gender','fever','cough','city']).values
x_train_age
```

```python
print(x_train_age.shape)
print(x_train_fever.shape)
print(x_train_gender_city.shape)
print(x_train_cough.shape)
```

```python
x_train_transform = np.concatenate((x_train_age,x_train_fever,x_train_gender_city,x_train_cough),axis = 1)
x_test_transfrom = np.concatenate((x_test_age, x_test_fever,x_test_gender_city,x_test_cough),axis = 1)
x_train_transform
```

# Column Transformer
## Transforming all the columns all at once 

```python
from sklearn.compose import ColumnTransformer
```

```python
# object      = operation        (list of all operations, rest remaining)
# transformer = ColumnTransformer(transformers=[],remainder = 'passthrough')
transformer = ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(),['fever']),
    ('tnf2', OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3', OneHotEncoder(sparse_output=False, drop = 'first'),['gender','city'])
],remainder = 'passthrough')
#remainder [passthrough = dont tough just pass, drop = remove]
```

```python
transformer.fit_transform(x_train).shape
```
