## Univarient Analysis

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('train.csv')
df
```

## Categorial Data

### Countplot

```python
# sns.countplot(df['Embarked'])
df['Embarked'].value_counts().plot(kind='bar')
```

### PieChart 
To get the data visualization in %


```python
df['Survived'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```
